```python
dfx = movies_data.loc[0:4, ['popularity','vote_average']]
plt.scatter(x= dfx['popularity'], y= dfx['vote_average'])
plt.xlabel("popularity")
plt.ylabel("vote_average")
plt.title("Movie-Recomm-Scatter -Plot")
plt.legend(loc="best", labels=["popularity-vote_average"])
plt.grid(True)
# plt.xlim(50, 80)
# plt.ylim(0, 2.5)
plt.show()
```
